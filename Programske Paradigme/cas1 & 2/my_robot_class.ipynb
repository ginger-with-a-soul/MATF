```python
import random


name = str(input("Tell me how to adress you kind stranger: "))
   
print(f"Hello {name}!. I will now generate a random number between [0, 100]. Can you guess my number?")

number = random.randint(0, 100)
tries = 0

while True:
    try:
        tries += 1
        x = int(input("Your guess: "))
    except TypeError as e:
        print("You didn't even enter an integer, moron!", e)
        exit(1)
        
    if x == number:
        print(f"Success!\nNumber of tries: {tries}")
        break
    elif x < number:
        print("Nope, your guess is smaller than my number!\nTry again!")
    else:
        print("Nope, your guess is greater than my number!\nTry again!")
```

```python
class Stack:
    def __init_(self):
        self.stack = []
    
    def push(self, element):
        self.stack.append(element)
    
    def pop(self):
        try:
            return self.stack.pop()
        except IndexError as e:
            print("Stack is empty! ", e)
            return None
        
    def peek(self):
        try:
            return self.stack[-1]
        except IndexError as e:
            print("Stack is empty! ", e)
            return None
        
    def getAll(self):
        print(*self.stack, sep = " ")


s = Stack()
s.push(1)
s.push(2)
s.push(3)
s.getAll()
s.pop()
print(s)
print(s.peek())
```

```python
class Robot:
    
    number_of_instances = 0
    
    def __init__(self, name):
        Robot.number_of_instances += 1
        self.name = name
        print(f'{self.name} has been created!')
        
    def __del__(self):
        Robot.number_of_instances -= 1
        print(f'{self.name} has been destroyed!')
        
    def say_hello_to_my_little_friend(self):
        print(f'Hello, my name is {self.name}! Nice to meet you :) *robot noises*')
        
    @classmethod
    def how_many(hm):
        print(f'There\'s {hm.number_of_instances} of us')

        
r1 = Robot("R2-D2")
r1.say_hello_to_my_little_friend()
r2 = Robot("R3PO")

Robot.how_many()

del r1

Robot.how_many()
```
